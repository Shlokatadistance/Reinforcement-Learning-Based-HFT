```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/singularity/Super_Fast/q-trader/run-gpu-6-04/q-trader/profitvstime.csv", header=None, usecols=[0,1],names=['Time','Profit_Selling'])
#print(df['Profit_Selling'].max())
print(len(df))
```

```python
print(df['Profit_Selling'].tail(400).max())
temp = []
for i in range(1182,1582):
    #print(df['Time'][i] - df['Time'][i-1] )
    temp.append(df['Time'][i] - df['Time'][i-1])
temp.insert(0,temp[0])

for i in range(len(temp)):
    temp[i] = temp[i] / 100
print(max(temp))
```

```python
plt.plot(df['Profit_Selling'].tail(400))
plt.title('Profits')
```

Improved trading efficiency, more profitable trades and a reduction in the loss from trades. These are extremely encouraging. The best trade resulted in a profit of ~2.6 whereas previosly the maximum profit from a trade was ~1.6
