```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/Users/singularity/Super_Fast/q-trader/returns_time.csv", header=None, usecols=[1,2],names=['Balance','Time'])
temp = []
for i in range(1,len(df)):
    #print(df['Time'][i] - df['Time'][i-1] )
    temp.append(df['Time'][i] - df['Time'][i-1])
    #df['latency'] = df['Time'][i] - df['Time'][i-1]
#print(df['latency'])
print(temp)
```

```python
plt.plot(df['Balance'] / 10000, label = 'portfolio_balance')
plt.plot(temp,label='time b/w trades')
plt.legend(loc='upper right')
```

Capital growth is slow, but as the model progresses and train, it makes both efficient and inefficient trading decisions. In some situations[x-(40,60)], there are higher trading times observed, and these rise as the model continues to make decisions. But eventually, as the model starts to growth its trading returns, albeit at wafer-thin margins, the latency also reduces.
